```python
import os
import pandas as pd

# Ruta base donde se encuentran los archivos
ruta_base = r"C:\Users\USUARIO\OneDrive - Secretaria de Educacion del Distrito\Escritorio\SOYhENRY\PI_21\PI_2_DA_corregido\data_internet_ind"

# Obtener la lista de archivos en la carpeta
archivos = [f for f in os.listdir(ruta_base) if f.endswith('.csv')]

# Diccionario para almacenar los DataFrames
dataframes = {}

# Cargar cada archivo de la lista
for archivo in archivos:
    ruta_completa = os.path.join(ruta_base, archivo)

    try:
        # Cargar el archivo CSV
        df = pd.read_csv(ruta_completa)

        # Normalizar los nombres de las columnas
        df.columns = (
            df.columns
            .str.strip()   # Eliminar espacios en los nombres de las columnas
            .str.lower()   # Convertir a minúsculas
            .str.replace(" ", "_")  # Reemplazar espacios por guiones bajos
            .str.replace(r"[^a-zA-Z0-9_]", "", regex=True)  # Eliminar caracteres especiales
        )

        # Guardar en el diccionario con un nombre limpio
        nombre_sin_extension = os.path.splitext(archivo)[0]
        dataframes[nombre_sin_extension] = df

        print(f"✅ Archivo '{archivo}' cargado y normalizado exitosamente.")

    except Exception as e:
        print(f"⚠️ Error al cargar '{archivo}': {e}")

# Mostrar las primeras filas de cada DataFrame cargado
for nombre, df in dataframes.items():
    print(f"\n📌 **{nombre}** - Primeras filas:")
    display(df.head())

# Ahora puedes acceder a los DataFrames con:
# dataframes["nombre_del_archivo_sin_extension"]
```
